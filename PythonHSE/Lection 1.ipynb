```python
import pandas as pd
import seaborn as sns

%config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

```python
tables = pd.read_html("https://en.wikipedia.org/wiki/Moscow", header = 1)
```

```python
table = tables[6]

table
```

```python
table.drop(10, inplace = True)
```

```python
table.Year = table.Year.astype(float)
```

```python
table.rename(columns = {'Pop.':'Pop'}, inplace = True)

list(table)
```

```python
table.Pop = table.Pop.astype(float)
```

```python
table.plot(x = 'Year', y = 'Pop')
```

```python
sns.lmplot(x = 'Year', y = 'Pop', data = table)
```

```python
sns.lmplot(x = 'Year', y = 'Pop', data = table, order = 2)
```

```python
sns.lmplot(x = 'Year', y = 'Pop', data = table, order = 3)
```
